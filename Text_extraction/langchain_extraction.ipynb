```python
from pdfminer.high_level import extract_text
import PyPDF2
import json
import re

def extract_text_from_pdf(pdf_path):

    text = extract_text(pdf_path)
    return text

def process_text_by_font_size(text):
    cur_fs = None
    cur_text = ''
    snippets = []

    for line in text.split('\n'):

        # Extract font size information using a simple regular expression
        fs_match = re.search(r'\bfont-size:(\d+)px\b', line)

        if fs_match:
            fs = int(fs_match.group(1))

            if cur_fs is None:
                cur_fs = fs

            if fs == cur_fs:
                cur_text += line + '\n'
            else:
                snippets.append((cur_text.strip(), cur_fs))
                cur_fs = fs
                cur_text = line + '\n'

    # Add the last snippet
    snippets.append((cur_text.strip(), cur_fs))

    return snippets

def main(pdf_path, output_json_path):
    # Extract text from the PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    print("Extracted text:")
    print(pdf_text)

    # Process text based on font size
    snippets = process_text_by_font_size(pdf_text)

    print(snippets)

    # Create a list of dictionaries for JSON serialization
    json_data = [{'text': text, 'font_size': font_size} for text, font_size in snippets]

    # Save JSON to a file
    with open(output_json_path, 'w') as json_file:
        json.dump(json_data, json_file, indent=2)

if __name__ == "__main__":
    # Replace 'your_pdf_path.pdf' with the actual path to your PDF file
    pdf_path = 'sustainability-report-fy2023.pdf'

    # Replace 'output.json' with the desired output JSON file path
    output_json_path = 'siemens_sustainability_report.json'

    main(pdf_path, output_json_path)



```

```python
import fitz  # PyMuPDF
import nltk
from nltk import sent_tokenize
import json
# from google.colab import files


nltk.download('punkt')

def extract_and_concatenate_by_heading(pdf_path):
    doc = fitz.open(pdf_path)
    pages_data = []

    for page_number in range(doc.page_count):
        page = doc[page_number]
        blocks = page.get_text("blocks")

        cur_heading = None
        cur_content = ''

        for block in blocks:
            text, font_size = block[4], block[5]

            # Assuming headings and subheadings are identified by their formatting
            if text.isupper():
                cur_heading = text
            elif text.istitle() and cur_heading:
                cur_content += text + ' '
            else:
                cur_content += text + ' '

        # Split content into sentences using NLTK
        sentences = sent_tokenize(cur_content.strip())

        # Append data for the current page
        pages_data.append({'page_number': page_number + 1, 'heading': cur_heading, 'content': cur_content.strip(), 'sentences': sentences})

    doc.close()
    return pages_data

def main(pdf_path, output_json_path):
    # Extract and concatenate text from the PDF
    pages_data = extract_and_concatenate_by_heading(pdf_path)

    # Save JSON to a file
    with open(output_json_path, 'w') as json_file:
        json.dump(pages_data, json_file, indent=2)

if __name__ == "__main__":

    pdf_path = 'sustainability-report-fy2023.pdf'


    output_json_path = 'siemens_sustainability_report.json'

    main(pdf_path, output_json_path)
```
